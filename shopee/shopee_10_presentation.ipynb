```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.chrome.service import Service
from time import sleep
from bs4 import NavigableString
import random
from random import randint
```

```python
#need
import time
import pandas as pd
df_name=pd.read_excel('20220511_tw.xlsx')
prefix='https://shopee.tw/mall/search?keyword='
sort_price='&order=asc&page=0&sortBy=price'

    
item_price=[]
item_name=[]
search_results=[]
list_index_of_min_price=[]
min_price_url=[]
for i in df_name['keyword']:
    
    
    print(i)
    base_url=prefix+i+sort_price

    
    chrome_options.add_argument('disable-notifications')
    
    chrome_options.add_argument('--disable-infobars')
    chrome_options.add_argument('start-maximized')
    chrome_options.add_argument('user-data-dir=C:\Program Files\Google\Chrome\Application')
    # To disable the message, "Chrome is being controlled by automated test software"
    chrome_options.add_argument("disable-infobars")
    # Pass the argument 1 to allow and 2 to block
    chrome_options.add_experimental_option("prefs", { 
        "profile.default_content_setting_values.notifications": 2
        })
        # invoke the webdriver
    path =  r'D:/Users/fungc1/Documents/chromedriver.exe'
    #s = Service(path)
    browser = webdriver.Chrome(path)
    browser.get(base_url)
    browser.maximize_window()
    delay = random.randint(2,4)
    

    
    try:
        WebDriverWait(browser, delay)
        #print ("Page is ready")
        time.sleep(random.randint(2,4))
        html = browser.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
        #print(html)
        soup = BeautifulSoup(html, "html.parser")


        if len(soup.findAll('div', class_='row shopee-search-item-result__items'))!=0:
            
            
            y = 1000
            for timer in range(0,10):
                browser.execute_script("window.scrollTo(0, "+str(y)+")")
                y += 1000  
                time.sleep(random.randint(2,5))
           
            
            html = browser.page_source #will return the html code with all products
            soup = BeautifulSoup(html, 'lxml')
            
            item_c_list=[]
            items_total=len(soup.find('div', class_='row shopee-search-item-result__items'))
            print(items_total)
            search_results.append(items_total)
            
            
            #for item_c in soup.findAll('span', class_='ZEgDH9'):
            for price_item in soup.findAll('div', class_='vioxXd pw1xTt rVLWG6'): 
                item_c=price_item.find('span', class_='ZEgDH9')
                item_c_int=int(item_c.text.replace(',',''))
                item_c_list.append(item_c_int)
            print(item_c_list)
            item_c_min=min(item_c_list)
            print(item_c_min)
            item_price.append(item_c_min)
                
            index_of_min_price = item_c_list.index(item_c_min)
            print(index_of_min_price)
            list_index_of_min_price.append(index_of_min_price)
            #item_price.append(item_c_min)
                
                #list_index_of_min_price.append(index_of_min_price)
            #print(index_of_min_price)
                
            
                    # find_all() returns an array of elements. 
                    # We have to go through all of them and select that one you are need. And than call get_text()
            for item_n in soup.findAll('div', class_='ie3A+n bM+7UW Cve6sh')[index_of_min_price]:
                #print(item_n.get_text)
                print(item_n)
                
                item_name.append(item_n)
                
            
            all_items_href=browser.find_elements_by_xpath('//a[@data-sqe="link"]')
            min_price_link_item=all_items_href[index_of_min_price]
            url=min_price_link_item.get_attribute('href')
            print(url)
            
            min_price_url.append(url)
            
            
            #for item in all_items_href:
                #url=item.get_attribute('href')
                #print(url)
               
               # min_price_url.append(url)
                        
                        
                        #a_href=soup.find("a",{"class":"data-sqe"}).get("href")
                        
                    
                            

                        

                    # find the price of items
           
            browser.close()
        else:
            print("no search result")
            print("N/A")
            print("N/A")
            print("N/A")
            print("N/A")
            
            item_price.append("N/A")
            item_name.append("N/A")
            search_results.append(0)
            list_index_of_min_price.append("N/A")
            min_price_url.append("N/A")
            browser.close()

        #break # it will break from the loop once the specific element will be present. 
    except TimeoutException:
        print ("Loading took too much time!-Try again")
```

```python
#need2
df_result = pd.DataFrame(list(zip(df_name['keyword'], search_results, list_index_of_min_price, item_name, item_price, min_price_url)),
               columns =['Keyword', 'No of Search Result', 'Index', 'Item Name', 'Item price', 'Min Price URL'])
#df_result['keyword']=df_name['keyword']

df_result.to_excel('taiwan_shopee_20220512_result10.xlsx', index=False)
```
